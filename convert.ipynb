```python
from lxml import etree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.collections import PatchCollection
from mpl_toolkits.basemap import Basemap
from shapely.geometry import Point, Polygon, MultiPoint, MultiPolygon, shape
from descartes import PolygonPatch
import mpl_toolkits.basemap.pyproj as pyproj
import fiona
from itertools import chain
```

```python
tree = etree.parse("data/plaques_20130119.xml")
root = tree.getroot()

output = dict()
output['raw'] = []
output['crs'] = []
output['lon'] = []
output['lat'] = []

for each in root.xpath('/openplaques/plaque/geo'):
    # check what we got back
    output['crs'].append(each.get('reference_system', None))
    output['lon'].append(each.get('longitude', None))
    output['lat'].append(each.get('latitude', None))
    # now go back up to plaque
    r = each.getparent().xpath('inscription/raw')[0]
    if isinstance(r.text, str):
        output['raw'].append(r.text.lstrip().rstrip())
    else:
        output['raw'].append(None)

df = pd.DataFrame(output)
df = df.replace({'raw': 0}, None)
df = df.dropna()
df.to_csv('data/output.csv')
df_map = pd.DataFrame({
    'poly': [shape(pol['geometry']) for pol in fiona.open('../pycity/data/ldn_ward_boundary/england_caswa_2001.shp')],
    })
wards_mp = MultiPolygon([p for p in df_map['poly']])
```

```python
plt.clf()
fig = plt.figure()
fig.set_size_inches(7.22, 5.25)
ax = fig.add_subplot(111, axisbg='w', frame_on=False)
extra = 0.01

# bounds need to be reprojected
wgs84 = pyproj.Proj("+init=EPSG:4326")
osgb36 = pyproj.Proj("+init=EPSG:27700")
bds = wards_mp.bounds
ll = pyproj.transform(osgb36, wgs84, bds[0], bds[1])
ur = pyproj.transform(osgb36, wgs84, bds[2], bds[3])
coords = list(chain(ll, ur))
w, h = coords[2] - coords[0], coords[3] - coords[1]
mid = pyproj.transform(osgb36, wgs84, wards_mp.centroid.xy[0], wards_mp.centroid.xy[1])

m = Basemap(
    projection='tmerc',
    lon_0 = mid[0][0],
    lat_0 = mid[1][0],
    ellps = 'WGS84',
    llcrnrlon=coords[0] - extra * w,
    llcrnrlat=coords[1] - extra + 0.01 * h,
    urcrnrlon=coords[2] + extra * w,
    urcrnrlat=coords[3] + extra + 0.01 * h,
    lat_ts=0,
    resolution='i')
m.readshapefile('../pycity/data/ldn_ward_boundary/england_caswa_2001_n', 'london', color='none', zorder=2)
to_dump = (Point(float(x), float(y)) for x, y in zip(
    df.lon,
    df.lat))
mp = MultiPoint([p for p in to_dump])

dev_x, dev_y = m(
    [point.x for point in mp],
    [point.y for point in mp])
dev = m.scatter(
    dev_x, dev_y, 7.5, marker='o', lw=.5,
    facecolor='#33ccff', edgecolor='w', alpha=0.8,
    antialiased=True, label='KCL AQSN NO$_2$ sensors', zorder=3)
# plot boroughs
patches = []
for xy, info in zip(m.london, m.london_info):
#     if info['AREA_CODE'] == 'LBO':
    patches.append(
        PolygonPatch(Polygon(xy), fc='#555555', ec='w', lw=.5, alpha=1., zorder=4))
ax.add_collection(PatchCollection(patches, match_original=True))

plt.xticks(())
plt.yticks(())
plt.title("Plaque Locations, London")
plt.tight_layout()
plt.savefig('data/london_plaques.png', dpi=300, alpha=True)
plt.show()
```

```python
def make_multipoint(df, reproject=False):
    """
    Return a Shapely Multipoint from a DataFrame
    Assumes that longitude and latitude columns exist
    """
    # we don't want any devices missing data
    # TODO: we need better data validation
    df[['lon', 'lat']] = df[['lon', 'lat']].astype(float)
    tpl = zip(df['lon'], df['lat'])
    if reproject:
        x, y = pyproj.transform(wgs84, osgb36, list(df['lon']), list(df['lat']))
        tpl = zip(x, y)
    return MultiPoint(tpl)

def colorbar_index(ncolors, cmap, labels=None, **kwargs):
    """
    This is a convenience function to stop you making off-by-one errors
    Takes a standard colourmap, and discretizes it,
    then draws a color bar with correctly aligned labels
    """
    cmap = cmap_discretize(cmap, ncolors)
    mappable = cm.ScalarMappable(cmap=cmap)
    mappable.set_array([])
    mappable.set_clim(-0.5, ncolors+0.5)
    colorbar = plt.colorbar(mappable, **kwargs)
    colorbar.set_ticks(np.linspace(0, ncolors, ncolors))
    colorbar.set_ticklabels(range(ncolors))
    if labels:
        colorbar.set_ticklabels(labels)

def cmap_discretize(cmap, N):
    """
    Return a discrete colormap from the continuous colormap cmap.
    
        cmap: colormap instance, eg. cm.jet. 
        N: number of colors.
    
    Example
        x = resize(arange(100), (5,100))
        djet = cmap_discretize(cm.jet, 5)
        imshow(x, cmap=djet)

    """
    if type(cmap) == str:
        cmap = get_cmap(cmap)
    colors_i = concatenate((linspace(0, 1., N), (0., 0., 0., 0.)))
    colors_rgba = cmap(colors_i)
    indices = linspace(0, 1., N + 1)
    cdict = {}
    for ki,key in enumerate(('red','green','blue')):
        cdict[key] = [(indices[i], colors_rgba[i - 1, ki], colors_rgba[i, ki]) for i in xrange(N + 1)]
    return matplotlib.colors.LinearSegmentedColormap(cmap.name + "_%d" % N, cdict, 1024)

def quantile(column):
    q = pd.qcut(column, 4)
    return len(q.levels) - q.labels
```

```python
points_ldn = make_multipoint(df, reproject=True)
```

```python
df_map['patches'] = df_map['poly'].map(lambda x: PolygonPatch(x, ec='#555555', lw=0.2, alpha=1., zorder=1))
df_map['count'] = df_map['poly'].map(lambda x: int(1. * len(filter(x.contains, points_ldn))))
df_map['area'] = df_map['poly'].map(lambda x: x.area)
df_map['density'] = df_map['count'] / df_map['area']
```

```python
df_map['density'] = df_map['density'].replace(to_replace=0.00000000e+00, value=np.nan)
categorical, levels = pd.cut(df_map['density'], 7, retbins=True)
labels = []
for l in levels:
    if l < 0:
        labels.append('None')
    else:
        labels.append("<= %0.7f/$ft^2$" % l)
```
